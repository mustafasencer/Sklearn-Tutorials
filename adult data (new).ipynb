```python
todummy_list = ["workclass", "education", "marital-status", "occupation", 
               "relationship", "race", "gender", "native-country"]
```

```python
def dummy_df(df, todummy_list):
    for x in todummy_list:
        dummies = pd.get_dummies(df[x], prefix = x, dummy_na=False)
        df = df.drop(x, axis = 1)
        df = pd.concat([df, dummies], axis = 1)
    return df
```

```python
X = dummy_df(X, todummy_list)
X.head()
```

# Handling missing Data
- Models can not handle missing data 

- Simplest solution 
    Remove obsrevations / features that have missing data

- But, removing missing data  can introduce a lot of issues
    Data is randomly missing: potentially lose a lot of your data
    Data is non-randomly missing: in addition to losing data, we are also introducing potential biases 
    Usually, this is a poor solution
- An alternative solution is to use imputation
    Replace missing value with another value
    Strategies: mean, median, highest frequency value of given feature

# More Data Exploration
- A large potion of the pre-modeling and modeling workflow can be generalized and automated 
- But understanding the problem, domain, and data is extremely important for building high performance models
- This section covers some tools used for exploring your data to make smarter decisions

```python
def find_outliers_tukey(x):
    q1 = np.percentile(x, 25)
    q3 = np.percentile(x, 75)
    iqr = q3 - q1 
    floor = q1 - 1.5*iqr
    ceiling = q3 + 1.5*iqr
    outlier_indices = list(x.index[ (x < floor) | (x > ceiling)])
    outlier_values = list(x[outlier_indices])
    return outlier_indices, outlier_values
```

```python
tukey_indices, tukey_values = find_outliers_tukey(X["age"])
np.sort(tukey_values)
len(tukey_indices)
```

# Distribution of Features
    - A histogram is a simple representation of the distribution of values for a  given feature
    - X-axis represents value bins and y-axis represents the frequency of an observation falling into that bin
    - It is also intresting to look at distributions broken up by outcome categories

```python
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
def plot_histogram(x):
    plt.hist(x,bins = 30, color = "blue", alpha = .5, lw = .3, edgecolor = "k")
    plt.title("Histogram of '{var_name}'".format(var_name = x.name))
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()
```

```python
plot_histogram(X["age"])
```

```python
def plot_histogram_dv(x, y):
    plt.hist(list(x[y == 0]) , color = "blue", alpha = .6, edgecolor = "k", 
            label = "DV = 0")
    plt.hist(list(x[y == 1]), color = "yellow", alpha = .4, edgecolor = "k",
            label = "DV = 1")
    plt.title("Histogram of '{var_name}' by DV category".format(var_name = x.name))
    plt.xlabel("value")
    plt.ylabel("frequency")
    plt.legend(loc = "best")
    plt.show()
```

```python
plot_histogram_dv(X["age"], y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

def find_model_perf(X_train, Y_train, X_test, Y_test):
    model = LogisticRegression()
    model.fit(X_train, Y_train)
    y_hat = [ x[1] for x in model.predict_proba(X_test)]
    
    auc = roc_auc_score(Y_test, y_hat)
    return auc
```
